```python
#swap the numbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
def swap_temp(a, b):
    a, b = b, a
    return a, b

def swap_arithmetic(a, b):
    a = a + b
    b = a - b
    a = a - b
    return a, b

def swap_xor(a, b):
    a = a ^ b
    b = a ^ b
    a = a ^ b
    return a, b

print("\nChoose Swapping Method")
print("1. Temporary Variable Method")
print("2. Arithmetic Method")
print("3. XOR Method")

choice = int(input("Enter your choice: "))

methods = {
    1: swap_temp,
    2: swap_arithmetic,
    3: swap_xor
}

if choice in methods:
    a, b = methods[choice](a, b)
    print("\nAfter Swapping:")
    print("First Number =", a)
    print("Second Number =", b)
else:
    print("Invalid Choice!")
```
