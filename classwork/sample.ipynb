```python
#Functions
def greet(name):
    return f"Hello, {name}!"
print(greet("Alice"))

#here the function is defined to greet a person by their name. It takes a parameter 'name' and returns a greeting message. When we call the function with "Alice", it will return "Hello, Alice!"
```

```python
#Arguments and its types
#1. Positional Arguments: These are the most common type of arguments. They are passed to a function in the order they are defined. For example:
def greet(name, age):
    return f"Hello, {name}! You are {age} years old."
print(greet("Alice", 25))
```

```python
#2 Default Arguments: These are arguments that have a default value. If the caller does not provide a value for these arguments, the default value will be used. For example:
def greet(name, age=30):
    return f"Hello, {name}! You are {age} years old."
print(greet("Bob"))
```

```python
#*args
def add(*numbers):
    print(sum(numbers))
add(1, 2, 3, 4, 5)    
```

```python
#**kwargs
def Profile(**data):
    for key, value in data.items():
        print(f" {key}: {value}")
Profile(name="Sameer", Role="Trainee", city="Hyderabad")    
```

```python
#Enclosing
def outer_function():
    x = "Hello"
    def inner_function():
        print(x)
    inner_function()
outer_function()    
```

```python
#Closures
def outer(x):
    def inner(y):
        return x + y
    return inner
add=outer(5)
print(add(10))
```

```python
#Non local Key word
def outer():
    count = 0
    def inner():
        nonlocal count
        count += 1
        print(count)
    inner()
outer() 
```
